```python
# Loading the dataset to work on
import pandas as pd
df = pd.read_csv("/content/shrink_ray_dataset.csv")
df.shape
```

```python
# Splitting the dataset columns to train the model with
x = df['Power']
y = df['Shrinkage']
```

```python
# Taking random elements from the splitted parameters to be used as a training set and the rest as testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state = 3)
```

```python
# preparing the sets to be in the form of 2d array to be suitable for Linear Regression
import numpy as np
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

```python
# Creating and Training a Linear Regression Model
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)

# Predicting the results of the trained Model with respect to the set it trained by
y_train_pred = lr.predict(x_train)

# Plotting the origninl set with respect to both the actual output and the output of the trained model
plt.scatter(x_train,y_train)
plt.scatter(x_train, y_train_pred, color="red")
plt.xlabel("Power (MW)")
plt.ylabel("Percentage (%)")
```

```python
# finding the output of a given set that the model hasn't dealt with before
y_test_pred = lr.predict(x_test)

# Plotting the origninl set with respect to both the actual output and the output of the trained model
plt.scatter(x_test,y_test)
plt.scatter(x_test, y_test_pred, color="red")
plt.xlabel("Power (MW)")
plt.ylabel("Percentage (%)")

```

```python
# The Linear Regression model will be in the form of Y = mX+b
m = lr.coef_
b = lr.intercept_
Y = 85.00
X = (Y-b)/m
print(f"GRU Should adjust the knob at {X[0][0]} to shrink the moon with a percentage 85%")
```
